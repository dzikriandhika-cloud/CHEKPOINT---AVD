OUTLIER

Penanganan Outliers
Outliers dideteksi menggunakan metode IQR untuk menghilangkan nilai ekstrem pada kolom total_pendapatan agar analisis lebih stabil.

```python
Q1 = df['total_pendapatan'].quantile(0.25)
Q3 = df['total_pendapatan'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['total_pendapatan'] >= lower) & (df['total_pendapatan'] <= upper)]

df = df.copy()
```

```python
df.shape
```

CONSTRUCT DATA
Konstruksi Data

Dilakukan pembuatan fitur baru yaitu pendapatan_per_jam yang diperoleh dari pembagian total_pendapatan dengan durasi_servis_per_jam untuk menganalisis efisiensi layanan.

```python
df['pendapatan_per_jam'] = df['total_pendapatan'] / df['durasi_servis_per_jam']
```
