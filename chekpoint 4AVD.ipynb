#Import Library & Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# load dataset

```python
df = pd.read_csv('/content/drive/MyDrive/avd praktikum/MFG10YearTerminationData.csv.zip')
```

# lihat data

```python
df.head()
```

# Bar Chart

# jumlah karyawan per departemen

```python
import matplotlib.pyplot as plt

dept = df['department_name'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(dept.index, dept.values)

plt.title('Top 10 Jumlah Karyawan per Departemen')
plt.xlabel('Departemen')
plt.ylabel('Jumlah Karyawan')

plt.xticks(rotation=45)
plt.show()
```

# Insight :
Grafik menunjukkan jumlah karyawan pada setiap departemen.


# Action :
Perusahaan dapat mengevaluasi distribusi karyawan agar beban kerja antar departemen lebih seimbang.

# Pie Chart

# presentase Gender Karyawan

```python
plt.figure(figsize=(6,6))

df['gender_full'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Persentase Gender Karyawan')
plt.ylabel('')

plt.show()
```

# Insight :
Grafik pie menunjukkan proporsi karyawan laki-laki dan perempuan di perusahaan.
# Action :
Data ini dapat digunakan untuk mengevaluasi keseimbangan gender dalam organisasi.

# line Chart

# Tren Jumlah Karyawan Per Tahun

```python
status_year = df.groupby('STATUS_YEAR')['EmployeeID'].count()

plt.figure(figsize=(10,6))
plt.plot(status_year.index, status_year.values, marker='o')

plt.title('Tren Jumlah Karyawan per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Karyawan')

plt.show()
```

#Insight :
Grafik menunjukkan perubahan jumlah karyawan dari tahun ke tahun.
#Action :
Jika terjadi penurunan jumlah karyawan, perusahaan perlu mengevaluasi penyebabnya.

#histogram

# Distribusi Umur Karyawan

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.hist(df['age'], bins=10, edgecolor='black')

plt.title('Distribusi Umur Karyawan')
plt.xlabel('Umur')
plt.ylabel('Jumlah Karyawan')

plt.show()
```

#Insight :
Histogram menunjukkan distribusi umur karyawan dalam perusahaan. Sebagian besar karyawan berada pada rentang usia produktif antara sekitar 25 hingga 55 tahun.
#Action :
Perusahaan dapat menggunakan informasi ini untuk merencanakan strategi pengembangan sumber daya manusia serta mempersiapkan regenerasi tenaga kerja di masa depan.

# Box Plot

#Lama Masa Kerja

```python
plt.figure(figsize=(8,6))

sns.boxplot(x=df['length_of_service'])

plt.title('Boxplot Lama Masa Kerja')

plt.show()
```

#Insight :
Boxplot membantu melihat median, sebaran, dan outlier masa kerja karyawan.
#Action :
Jika banyak karyawan dengan masa kerja pendek, perusahaan perlu meningkatkan retensi karyawan.

#Scatter Plot

#Hubungan Umur Dan Lama Kerja

```python
plt.figure(figsize=(8,6))

plt.scatter(df['age'], df['length_of_service'])

plt.title('Hubungan Umur dan Lama Kerja')
plt.xlabel('Umur')
plt.ylabel('Lama Kerja')

plt.show()
```

#Insight :
Scatter plot menunjukkan apakah karyawan yang lebih tua cenderung memiliki masa kerja lebih lama.
#Action :
Perusahaan dapat melihat pola loyalitas karyawan berdasarkan usia.

#Heatmap

#Korelasi Antar Variabel Numerik

```python
plt.figure(figsize=(8,6))

corr = df[['age','length_of_service']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Heatmap Korelasi Data')

plt.show()
```

#Insight :
Heatmap menunjukkan hubungan antar variabel numerik dalam dataset.
#Action :
Jika terdapat korelasi kuat, perusahaan dapat menggunakan variabel tersebut untuk analisis lebih lanjut.
