```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MFG10YearTerminationData.csv.zip', compression='zip')
df.head()
```

```python
df.columns
```

```python
df['Terminated'] = df['terminationdate_key'].apply(lambda x: 0 if x == '1/1/1900' else 1)
df['Terminated'].value_counts()
```

```python
df['Terminated'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Terminated', y='age', data=df)
plt.title("Age Distribution by Termination Status")
plt.show()
```

INSIGHT

Dari analisis data, terlihat bahwa tingkat attrition lebih tinggi pada karyawan dengan masa kerja yang masih rendah dan pendapatan yang lebih kecil.

REKOMENDASI

Perusahaan dapat memberikan perhatian lebih kepada karyawan dengan masa kerja rendah melalui program pelatihan, peningkatan engagement, serta evaluasi sistem kompensasi untuk mengurangi tingkat attrition.

KESIMPULAN

Berdasarkan analisis, masa kerja dan pendapatan berpengaruh terhadap tingkat attrition. Karyawan dengan masa kerja rendah dan pendapatan kecil cenderung lebih sering keluar.
